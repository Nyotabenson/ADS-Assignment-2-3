### FACTORS CAUSING ATTRITION
### 1.1 Personal factors:-
 a. Gender
 
 b. Age
 
c. Maritals status

Plotting the above factors in relation to attrition to understand their impact

```python
plt.figure(figsize=(12,3))

plt.subplot(1,3,1)
sns.countplot(data=df, x="Gender", hue="Attrition")
plt.title("Gender in relation\nto Attrition")

plt.subplot(1,3,2)
sns.countplot(data=df, x="Age", hue="Attrition")
plt.xticks(np.arange(10,60,10))
plt.title("Age in relation\nto Attrition")

plt.subplot(1,3,3)
sns.countplot(data=df, x="MaritalStatus", hue="Attrition")
plt.title("MaritalStatus in relation\nto Attrition")

plt.show()
```

### obs: 
 from the observation Male gender has the highest rate of attrition compared to female gender
 
 midle age employees has high attrition rate
 
 Employees who are single has high attrition than Married and divorced, while as the Married has high attrition than the divorced.
